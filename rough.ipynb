```python
import numpy as np
from tensorflow.keras.utils import Sequence

class DataGenerator(Sequence):
    def __init__(self, X_data, y_data, batch_size, shuffle=True):
        self.X_data = X_data
        self.y_data = y_data
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.indices = np.arange(len(X_data))
        self.on_epoch_end()  # Initialize shuffle

    def __len__(self):
        """Defines the number of batches per epoch"""
        return int(np.floor(len(self.X_data) / self.batch_size))

    def __getitem__(self, index):
        """Generates one batch of data"""
        batch_indices = self.indices[index * self.batch_size:(index + 1) * self.batch_size]
        X_batch = self.X_data[batch_indices]
        y_batch = self.y_data[batch_indices]
        return X_batch, y_batch

    def on_epoch_end(self):
        """Shuffle indices after each epoch if enabled"""
        if self.shuffle:
            np.random.shuffle(self.indices)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import backend as K
from tensorflow.keras import metrics

def f1_score(y_true, y_pred):
    y_pred = tf.round(y_pred)  # Round predictions to 0 or 1
    tp = K.sum(K.cast(y_true * y_pred, 'float'), axis=1)  # True positives
    fp = K.sum(K.cast((1 - y_true) * y_pred, 'float'), axis=1)  # False positives
    fn = K.sum(K.cast(y_true * (1 - y_pred), 'float'), axis=1)  # False negatives

    precision = tp / (tp + fp + K.epsilon())
    recall = tp / (tp + fn + K.epsilon())
    f1 = 2 * precision * recall / (precision + recall + K.epsilon())
    return K.mean(f1)

# Define the model
model = keras.Sequential([
    keras.layers.Dense(1024, input_shape=(1024,), activation='relu'),
    keras.layers.Dense(512, activation='relu'),
    keras.layers.Dense(1373, activation='sigmoid')  # Use sigmoid for multi-label classification
])

# Compile the model with binary_crossentropy for multi-label classification
model.compile(
    optimizer='Adam',
    loss='binary_crossentropy',  # Use binary_crossentropy for multi-hot encoded labels
    metrics=[metrics.Precision(), metrics.Recall()]
)
early_stopping = EarlyStopping(monitor='val_loss', patience=5)

# Train the model
# model.fit(xs_train, ys_train, epochs=2)
model.summary()
```
